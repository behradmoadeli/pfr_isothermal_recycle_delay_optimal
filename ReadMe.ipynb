Initializing the run:

```python
pars_list, default_par = create_custom_pars_list("runs/readme_run/pars_list.csv")
r = [-100, 20, 45]
i = [0, 360, 120]
f = [(-100, 25), 300]
```

Running for the default parameters:

```python
df, label, metadata = find_eig(par=default_par, guess_range_real=r, guess_range_imag=i)
save_dataframe_to_csv(df, label, "runs/readme_run/CSV", metadata)
plot_single_df(df, label, metadata, filter=False)
display(df)
```

Running for variations in default parameters:
